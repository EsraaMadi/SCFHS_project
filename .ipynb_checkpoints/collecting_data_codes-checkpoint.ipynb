```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

import pandas as pd # I'll be using pandas to store & save the results
import numpy as np
```

```python
# Specifying incognito mode as you launch your browser [OPTIONAL]
option = webdriver.ChromeOptions()
option.add_argument("--incognito")
```

```python
# Create new Instance of Chrome in incognito mode
browser = webdriver.Chrome(executable_path='/Users/saud/Documents/Projects/WebScraping/chromedriver', 
                           chrome_options=option)
```

```python

# Go to desired website
browser.get("https://services.amana-md.gov.sa/eservicesite/Inq/DeathInquiry.aspx")

# 2nd: I did the query manually from the year 1340 to the date (1439/12/30)
# this makes the code easier to read 
# this step can be automated
```

```python
# Extract the target elemnts from the table 
# find_elements_by_xpath - Returns an array of selenium objects.
# by checking the xpath of the desired data using inspects 
data = browser.find_elements_by_xpath('//*[@id="ctl00_ContentPlaceHolder1_dgDeath"]/tbody/tr')

```

```python
# check data 
for line in data[2:-1]: # the first and last elemnts are indexes, the second elemnt is the column name
    print(line.text.split(' ')[-1])# since we are only intersted in Date of burial  
```

```python
# create an empty series to save the data 
data = pd.Series(name='date',dtype='object')
data.head()
```

```python
# converting to dataframe
data.head().to_frame()
# then we can save it into a csv
```

```python
# creating an empty csv file to dump the result 
data.to_frame().to_csv('data/data_text.csv', mode='a',header=False,index=False)
```

```python
%%time

import time
# loop over pages 

# we will be extracting the data untill it finished
pages_counter = 0
pages = [page for page in range(1,21)] + [21] # this 21 is to go to the next 20 pages --> ...

try:
    while True: # because we don't know when pages will end
        for page in pages:#starts from 2 to 20
            
            


            # extract & save data for each pages 
            # extract dates 
            data_path = browser.find_elements_by_xpath('//*[@id="ctl00_ContentPlaceHolder1_dgDeath"]/tbody/tr')
            print('number of dates for this page:', len(data_path[2:-1]))
            #save them into a list
            dates = [line.text.split(' ')[-1] # taking the date only for counts
                     for line in data_path[2:-1]] 
            
            # create an empty series to save the data 
            data = pd.Series(data=dates,
                             name='date',
                             dtype='object')

            # append the result in a csv file
            data.to_frame().to_csv('data/data.csv', mode='a',header=False,index=False)
            
            # finding the next page clicker using xpath as well
            page_clicker = browser.find_elements_by_xpath('//*[@id="ctl00_ContentPlaceHolder1_dgDeath"]/tbody/tr[78]/td/a[{}]'.format(page))[0]

            # next page 
            page_clicker.click()
            time.sleep(5)
            
            pages_counter += 1
            print(pages_counter) # to see the progress
            
        if pages.count(1)==1:
            # we have to do this, removing 1, starting from the second round
            pages = pages.remove(1) + [21] # this 21 is to go to the next 20 pages --> ... 
except Exception as e:
    print(e)
    print('Done extracting')
        
    
```
